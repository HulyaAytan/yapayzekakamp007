## Text Preprocessing 

```python
df['text']= df['subject'] + " " + df['title'] + " " + df['text']
del df['title']
del df['subject']
del df['date']
```

```python
df.head()
```

```python
df['text'][0]
```

```python
df["text"]=df["text"].str.lower() #küçük harfe çevrildi
df["text"]=df["text"].str.replace("[^\w\s]","") #noktalama işaretlerini kaldırır
df["text"]=df["text"].str.replace("\d+","") #d digit rakamları kaldırır
df["text"]=df["text"].str.replace("\n"," ").replace("\r","") #Satır sonu,\n enter a basılmışsa \r
```

## Wordclouds

```python
def wc(data,bgcolor): 
    plt.figure(figsize=(10,10))
    wc=WordCloud(background_color=bgcolor, stopwords=STOPWORDS)
    wc.generate(' '.join(data))
    plt.imshow(wc) 
    plt.axis("off")
```

## Apply Model

```python
#define x,y
x=df["text"] 
y=df["class"]
```

```python
vect=CountVectorizer(max_features = 5000,stop_words="english")
```

```python
x=vect.fit_transform(x).toarray()
```

```python
# split to train and test data
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=.15)
```

```python

from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier


mb=MultinomialNB()
D=DecisionTreeClassifier()
KN=KNeighborsClassifier() 
R=RandomForestClassifier()
XGB=XGBClassifier()

classifiers = [mb, D, KN, R, XGB]


classifier_names = ['MultinomialNB', 'DecisionTreeClassifier', 'KNeighborsClassifier', 'RandomForestClassifier', 'XGBClassifier']


df = pd.DataFrame(columns=['classifier', 'accuracy'])


for i, clf in enumerate(classifiers):
    pred = clf.fit(x_train, y_train).predict(x_test)
    score = accuracy_score(pred, y_test)
    df.loc[i] = [classifier_names[i], score]
```

```python
df.sort_values(by='accuracy', ascending=False, inplace=True)
df
```
