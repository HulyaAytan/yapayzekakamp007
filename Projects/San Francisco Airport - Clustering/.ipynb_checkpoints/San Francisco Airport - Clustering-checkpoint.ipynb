```python
df1=df[['Operating Airline','Passenger Count']]
```

```python
df1
```

```python
df1=pd.get_dummies(df1,drop_first=True)
```

```python
df1
```

```python
model=KMeans(n_clusters=4)
```

```python
tahmin=model.fit(df1)
```

```python
tahmin=model.predict(df1)
```

```python
tahmin
```

```python
df1["cluster"]=tahmin
```

```python
df1.head()
```

```python
silhouette_score(df1,tahmin)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,20))
plt.scatter(x=df['Operating Airline'],y=df['Passenger Count'],c=tahmin)
plt.xlabel("Operating Airlines")
plt.ylabel("Passenger Count")
plt.xticks(rotation=90)
plt.colorbar()
plt.grid()
```

```python
allscore=[]
allclusters=[]
uzakliklar=[]
def clust(kactane):
    
    for i in range(1,kactane+1):
        i=i+1
        model=KMeans(n_clusters=i)
        model=model.fit(df1)
        tahmin=model.predict(df1)
        score=silhouette_score(df1,tahmin)
        uzaklik=model.inertia_ 
        print("Cluster",i,"score",score)
        allscore.append(score)
        allclusters.append(i)
        uzakliklar.append(uzaklik)
clust(10)
```

```python
plt.xlabel("k")
plt.ylabel("uzaklÄ±klar")
plt.plot(allclusters,uzakliklar)
```
