### Feature Engineering

```python
df["zipcode"]=df["zipcode"].astype("object") #zipcode yazıya döndü
```

```python
df["age"]=(2022-df["yr_built"])**2
```

```python
df["Renovated"]=(2022-df["yr_renovated"])**2
```

```python
df["bedrooms"]=df["bedrooms"]**2
```

```python
df['bathrooms']=df['bathrooms']**2
```

```python
outliers = df.quantile(.97)
df = df[(df['price']<outliers['price'])]
df = df[(df['bedrooms']<outliers['bedrooms'])]
df = df[(df['sqft_living']<outliers['sqft_living'])]
```

```python
#sns.pairplot(df) #burada butun sutunların diğer sütunlara göre değişimini grafiklerle gösterdik.
```

### Modelling

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression #linear regression import ettik
```

```python
lr=LinearRegression()
```

```python
from sklearn.metrics import r2_score, mean_squared_error #rootmeansquare araba başına hatamız
```

```python
x=df[["sqft_living","grade","sqft_above","sqft_living15","bathrooms","view","sqft_basement","bedrooms","lat","waterfront","floors","zipcode","age","yr_renovated","condition"]] #Mileage domain bilgisi, basarı oranını artırmak için yazı olan verileri ekledik mare and trim
y=df[["price"]]
```

```python
 x=pd.get_dummies(x,drop_first=True)#herbir yazı değeri için yeni bir sütun açıyor drop_first=True birinci sütunu sildi
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
model=lr.fit(x_train,y_train)
```

```python
tahmin=model.predict(x_test)
```

```python
(mean_squared_error(tahmin,y_test))**(0.5)
```

```python
r2_score(tahmin,y_test) #sözcük içeren sütunları eklemek başarı oranını artırdı
```

```python
from sklearn.linear_model import Ridge,Lasso #aynı işlemi iki farklı modelle deniyoruz
```
