```python
import pandas as pd
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential #sıraya koymamızı sağlayacak
from tensorflow.keras.layers import Dense #bütün nöronları birbirine bağlayacak
```

```python
df=pd.read_csv("pima-indians-diabetes.csv")
```

```python
df.head() #hamile kadınların diyabet durumlarına bakıyor, 1 diyabet 0 diyabet değil(outcome).
```

```python
df.isnull().sum() #boş veri var mı bakıyoruz
```

```python
df.columns
```

```python
x=df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]
```

```python
y=df[["Outcome"]]
```

```python
#x=df.iloc[:,0:8] x ve y yi bu şekildede yazabiliyoruz
#y=df.iloc[:,8]
```

```python
x.shape #8 sütundan oluşuyor bu yüzden 1.sine 8 tane nöron koyuyoruz diğerlerine istediğimiz kadar
#veri sayısı 1000 den fazla olduğunda derin öğrenme güzel sonuç veriyor
```

```python
model=Sequential() #modelimizi tanımladık 
model.add(Dense(8,activation='relu'))#1. nöronun diğer nöronlara bağlanacağını söylüyoruz
model.add(Dense(12,activation='relu'))
model.add(Dense(4,activation='relu')) #birden fazla layer varsa relu
model.add(Dense(1,activation='sigmoid')) #cevabı evet hayırsa sigmoid.son nöronun aktivasyon kodunun 1 olması lazım,çünkü sonuncu evet yada hayır olacak
model.compile(loss="binary_crossentropy",optimizer="adam",metrics="accuracy")
#softmax bşrden fazla class varsa
```

```python
layer sayısını artırmak
nöron sayısını artırmak 
```

```python
model.fit(x,y,epochs=150,batch_size=10,verbose=1) #epoch, nöronlar arasında kaç defa gidip geleceğimizi belirliyoruz
#verbose aşağıda çalıştığını gösteriyor, batch_size bir seferde 10 satırlık veri alıyor
```

```python
model.summary()
```

```python
scores=model.evaluate(x,y) #loss hata
```

```python
scores[0]
```

```python
scores[1]
```

```python
#Early stopping - 150 epoch biz 150 epoch girdik ama ideal olan 75se, orada durması early stopping
```

```python
history=model.fit(x,y,epochs=150,validation_split=0.20,batch_size=10,verbose=1)
```

```python
import pandas as pd
pd.set_option("display.max_columns",100)
```

```python
df=pd.read_pickle("kc_house.pkl")
```

```python
#df.to_csv("kcdata_csv")
```

```python
df.head()
```

```python
df.price
```

```python
x=df.drop("price",axis=1) #price sütunu drop ediyor diğerlerini alıyor
y=df[["price"]]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
model=Sequential()
model.add(Dense(19,activation='relu')) #sütun sayısı olacak 81 i 19 yaptık deneme
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1)) #rakam tahmin ettiğimiz için 1 olmasını istiyoruz
model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x,y,validation_split=0.20,batch_size=128,epochs=100)
model.summary()
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=128,epochs=100)
model.summary()
```

```python
loss_df=pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

```python
tahmin=model.predict(x)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
r2_score(tahmin,y) #%80 üstü bekliyoruz
```

```python
(mean_squared_error(tahmin,y))**0.5 #100000 altı
```
